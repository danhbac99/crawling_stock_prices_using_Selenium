```python
url = 'https://quotes.vcbs.com.vn/a/exchange.html?symbol=HSX'
driver = webdriver.Chrome() 
driver.get(url)
```

```python
stocks = driver.find_elements(By.CLASS_NAME,"StockSymbolEven")
ticker_list = []
limit= 60
for stock in stocks[:limit]:
    ticker_list.append(stock.text)
print(ticker_list)
```

```python
Stock =[]
driver = webdriver.Chrome()
for symbol in ticker_list:
    url = 'https://dstock.vndirect.com.vn/lich-su-gia/'+ symbol
    driver.get(url)
    time.sleep(2)

    try:
        eclose_ads = driver.find_element(By.XPATH, "//button[@class='sc-bdfBwQ hoqISq sc-hKgILt gTLZXx reactour__close']").click()
    except NoSuchElementException:
        pass
    #print("No element found")
    start = driver.find_element(By.XPATH, "//input[@class='data-market__date data-market__from']")
    start.click()
    start.send_keys('01/01/2021')
    time.sleep(1)
    end = driver.find_element(By.XPATH, "//input[@class='data-market__date data-market__to']")
    end.click()
    end.send_keys('12/12/2021')
    time.sleep(1)
    find = driver.find_element(By.XPATH, "//button[@class='btn btn--primary btn--small']")
    find.click()
    time.sleep(1)
    for k in range(100):
        count = 1
        date = driver.find_elements(By.XPATH, "//table[@class='data-market__width hidden-mobile']/tbody/tr/td[1]")
        price = driver.find_elements(By.XPATH, "//table[@class='data-market__width hidden-mobile']/tbody/tr/td[6]")
        for i in range(len(date)):
            print(symbol + ':' + date[i].text + ':' +price[i].text)
            data = {
                'year': date[i].text,
                'symbol': symbol,
                'price': price[i].text
            }
            Stock.append(data)
        try:
            next = driver.find_element(By.XPATH, '//li[contains(@class, "page-item") and contains(@class,"active")]/following-sibling::li/a')
            next.click()    
            count = count + 1
            time.sleep(1)
        except:
            break

    print("Completed.....")
```

```python
df = pd.DataFrame(Stock)
df
```

```python
df = df.to_csv('stock1.csv', index=False)
```
